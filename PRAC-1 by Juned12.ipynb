```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create Dataset
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug"],
    "Region": ["North","South","East","West","North","South","East","West"],
    "Sales": [20000,22000,25000,24000,26000,27000,30000,32000],
    "Advertising_Spend": [3000,3200,3500,3300,3600,3800,4000,4200],
    "Profit": [5000,5500,6200,6000,6500,7000,7500,8000]
}

df = pd.DataFrame(data)
df
```

### Group Analysis

### VISULIZATION

```python
# sales trend
plt.plot(df["Month"], df["Sales"])
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Advertising vs Sales
plt.scatter(df["Advertising_Spend"], df["Sales"])
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.title("Advertising vs Sales")
plt.show()
```

```python
# Profit Distribution
plt.bar(df["Month"], df["Profit"])
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```
